```python
#-Q2-What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram

track_popularity = df.groupby('Track Name')['Popularity'].mean().reset_index()
track_popularity
```

```python
plt.figure(figsize=(10, 6))

# Loop over unique track names
for track in df['Track Name'].unique():
    sns.histplot(
        df[df['Track Name'] == track]['Popularity'],
        kde=True,
        label=track,
        alpha=0.5
    )

# Add plot details
plt.title('Distribution of Popularity by Track')
plt.xlabel('Popularity Score')
plt.ylabel('Frequency')
plt.legend(title='Track Name')

# Show plot
plt.show()
```

```python
#-Q4-Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot
df
```

```python
df['Artist'].value_counts()
```

```python
artist_track_count = df['Artist'].value_counts()
print("Number of Tracks for Each Artist:")
print(artist_track_count)

# Create a countplot for the number of tracks per artist
plt.figure(figsize=(18, 6))
sns.countplot(data=df, x='Artist', order=artist_track_count.index, palette='viridis')

# Add plot details
plt.title('Number of Tracks by Each Artist')
plt.xlabel('Artist')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
artist_popularity = df.groupby('Artist')['Popularity'].sum()
artist_popularity
```

```python
top_5_artists = artist_popularity.nlargest(5).index
top_5_artists
```

```python
top_5_artists_df = df[df['Artist'].isin(top_5_artists)]
top_5_artists_df
```

```python
average_popularity = top_5_artists_df.groupby('Artist')['Popularity'].mean()
average_popularity
```

```python
#-Q7_For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist
artist_popularity = df.groupby('Artist')['Popularity'].sum()
top_5_artists = artist_popularity.nlargest(5).index
top_5_artists_df = df[df['Artist'].isin(top_5_artists)]['Track Name']	

top_5_artists_df
```

```python
#-Q9- Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Artist', y='Duration (ms)', palette='Set2')

# Add titles and labels
plt.title('Track Duration Across Different Artists', fontsize=16)
plt.xlabel('Artist', fontsize=12)
plt.ylabel('Duration (ms)', fontsize=12)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a violin plot for the duration of tracks across different artists
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Artist', y='Duration (ms)', palette='Set2')

# Add titles and labels
plt.title('Track Duration Distribution Across Different Artists', fontsize=16)
plt.xlabel('Artist', fontsize=12)
plt.ylabel('Duration (ms)', fontsize=12)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
