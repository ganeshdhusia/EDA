```python
#-Q5-What are the most popular product names

average_ratings = df.groupby('product_name')['rating'].mean()

highest_rated_product = average_ratings.idxmax()
highest_average_rating = average_ratings.max()
print(f"the most popular product names is: {highest_rated_product} with a rating of {highest_average_rating:.2f}")
```
