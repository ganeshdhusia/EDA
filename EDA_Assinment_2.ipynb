```python
#_Q3_> What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?

df['Transmission']=df['Transmission'].str.replace ('Auto', '1').str.replace('Manual', '2')
```

```python
df['Transmission'] = df['Transmission'].astype (int)
```

```python
plt.figure(figsize=(15, 9))
sns.histplot(df['Transmission'], label='Transmission', kde=True)
sns.histplot(df['Price ($)'], label='Price ($)', kde=True)
plt.title('Histogram of Transmission and Price ($)')
plt.xlabel('Transmission')
plt.ylabel('price')
plt.legend()
plt.show()
```

```python
#-Q-9 How do car prices vary based on the customer’s annual income bracketE
df[['Annual Income','Price']]
```

```python
correlation = df['Annual Income'].corr(df['Price'])
print(f"Correlation between Price and Annual Income: {correlation:.2f}")


sns.scatterplot(x='Price', y='Annual Income', data=df, color='blue', marker='o')
plt.title('Price vs Annual Income')
plt.xlabel('Price')
plt.ylabel('Annual Income')
plt.grid(True)
plt.show()
```

```python
#_Q-10_V What are the top 5 car models with the highest number of sales, and how does their price distribution lookE

filtered_df = df[['Model','Price']].value_counts()[:5].reset_index()
filtered_df.columns = ['Model', 'Price', 'Count']
filtered_df = df.merge(filtered_df[['Model', 'Price']], on=['Model', 'Price'])
sns.boxplot(x='Model', y='Price', data=filtered_df, palette='pastel')
plt.title('Boxplot of Price for Top 5 Model-Price Combinations')
plt.xlabel('Model')
plt.ylabel('Price')
plt.show()
```
